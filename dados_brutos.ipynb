```python
#importando dados
import pandas as pd

name = "UTE Celso Furtado (Bahia - S F C) - 2019 - 05 - mensal_2019_06_03_11_17_01.dat"
#print(data1) se vc printar antes dos comandos abaixo entenderá pq..

data1 = pd.read_csv(name, header = 1) # o nome das colunas estão na primeira linha

data1 = data1[2:] # os valores começam a partir da 2 linha

data1.head()
```

```python
import datetime as dtime

data1['TIMESTAMP'] = pd.to_datetime(data1["TIMESTAMP"])

# Coluna 'Data'
data1['Data'] = data1['TIMESTAMP'].dt.strftime('%d/%m/%Y') #colocando no formato dia/mes/ano
# Coluna 'Hora'
data1['Hora'] = data1['TIMESTAMP'].dt.strftime('%H:%M:%S')
#seprando uma coluna só para dia
data1['dia_new'] = data1.TIMESTAMP.dt.day
#criando uma coluna de dia juliano
data1['julian_day'] = data1.TIMESTAMP.dt.dayofyear

#del data1['TIMESTAMP'] # n vou precisar mais agr vou ponderar a ideia de indexiar o dataframe pela coluna data ou hora
data1.head()
```

```python
data1['Hora'] = data1['Hora'] - dtime.timedelta(hours = 3)
# data1['Data'] = pd.to_datetime(data1['Data']) --  talvez
data1['TIMESTAMP'] = data1['TIMESTAMP'] - dtime.timedelta(hours = 3)

data1.head() # será perguntado ao user se o dado estará 3 horas a frente 
```

```python
#criando novas colunas
import numpy as np

data1['Hora_solar'] = 0 
data1['ET'] = 0
data1['Dia_angular'] = 0
data1['ang_hr'] = 0
data1['Hora_local'] = 0 
data1['Declinacao'] = 0
data1['Long ref'] = -45
data1['Longitude'] = -38.57034 #samira passou
data1['Latitude'] = -12.70328 #samira passou
################################################################################################################################

data1['Declinacao'] = 23.45 * np.sin(np.deg2rad((360/365)*(data1['julian_day'] + 284)))
 
data1['Hora_local'] = (data1.TIMESTAMP.dt.hour * 60 ) + 1 + data1.TIMESTAMP.dt.minute # depois criar uma função só pra isso

data1['Dia_angular'] = (2*np.pi) * (data1['julian_day'].astype(float)-1)/365

data1['ET'] = 229.18 * ((0.000075 + 0.001868*np.cos(np.deg2rad(data1['Dia_angular']))) - (0.032077*(np.sin(np.deg2rad(data1['Dia_angular'])))) - (0.014615*(np.cos(np.deg2rad(2*data1['Dia_angular']))))  - (0.04089*(np.sin(np.deg2rad(2*data1['Dia_angular'])))))

data1['Hora_solar'] = data1['Hora_local'] - (4*(data1['Long ref'] - data1['Longitude']) + data1['ET'])/60
                          
data1['ang_hr'] = (data1['Hora_solar'] - 12)*15  

data1.head()
```

## Calculando cosseno 

```python
# a equação que queremos descobrir é:
# # GH_Max = (Sa x 1.5 x (u*1.2) + 100 ) ------------------ expressão para calcular GH_Max
# para achar u teremos que
# u = cos(AZ) = cos(lat_local)*cos(decli_solar)*cos(ang_hora) + sen(lat_local)*sen(decli_solar)
# agr calculando o u teremos que

data1['Coseno_AZ']= 0

data1['Coseno_AZ']= np.cos(np.deg2rad(data1['Latitude']))*np.cos(np.deg2rad(data1['Declinacao']))*np.cos(np.deg2rad(data1['ang_hr'])) + np.sin(np.deg2rad(data1['Latitude']))*np.sin(np.deg2rad(data1['Declinacao'])) 

print("Se Az > 90, u é nulo")
data1.head()
```

## Calculando Sa

```python
# Sa = So.Eo
# sendo So uma constante = 1367
# e Eo= 1.000110 + 0.0334221*cos(diaangular) + 0.001280*sen(diaangular) + 0.000719*cos(diangular*2) + 0.000077sin(diaangular*2)

data1['Sa'] = 0
data1['So'] = 1367 #criando coluna So
data1['Eo'] = 0

# Equação de correção de excentricidade de orbita terrestre
data1['Eo'] = 1.000110 + (0.0334221*(np.cos(np.deg2rad(data1['Dia_angular'])))) + (0.001280*(np.sin(np.deg2rad(data1['Dia_angular'])))) + (0.000719*(np.cos(np.deg2rad((2*data1['Dia_angular']))))) + (0.000077*(np.sin(np.deg2rad((2*data1['Dia_angular'])))))

data1['Sa'] = data1['Eo'] * data1['So']

data1.head()
```

```python
data1['Coseno_AZ_Ang'] = np.rad2deg(np.arccos(data1['Coseno_AZ']))
######################################################## teste ################################################################

#criando coluna com U elevado a 1.2
data1.loc[data1['Coseno_AZ_Ang'] > 90 , '1.2_do_cosseno'] = 0
data1.loc[data1['Coseno_AZ_Ang'] < 90 , '1.2_do_cosseno'] = (data1['Coseno_AZ'])**(1.2)

# Criando coluna com U elevado a 0.2
data1.loc[data1['Coseno_AZ_Ang'] > 90 , '0.2_do_cosseno'] = 0
data1.loc[data1['Coseno_AZ_Ang'] < 90 , '0.2_do_cosseno'] = (data1['Coseno_AZ'])**(0.2)

data1.head(420) # deu certo
```
